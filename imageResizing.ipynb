```python
# resizing the image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('/content/pic_0009.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20, 20))

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)


img_scaled = cv2.resize(image, (256, 256), interpolation = cv2.INTER_AREA)

plt.subplot(1, 2, 2)
plt.title("resized")
plt.imshow(img_scaled)
```
