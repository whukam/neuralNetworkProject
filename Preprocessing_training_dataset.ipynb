```python
file_dataset = './gdrive/MyDrive/NNProDrive/neuralNetworkProject/dataset/'
file_Preprocessing = './gdrive/MyDrive/NNProDrive/neuralNetworkProject/Preprocessing/'


# iterate all files from a directory
for file_name in os.listdir(file_dataset):

    if file_name == 'desc':
        continue
    # getting full name for images
    source = file_dataset + file_name

    # getting destination full name
    destination = file_Preprocessing + file_name

    image = cv2.imread(source)
    #image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    #rescaling the image into deminsion of 256x256 size 
    img_scaled = cv2.resize(image,(256,256), interpolation = cv2.INTER_AREA)

    #sharpening the images 
    #Create our shapening kernel, we don't normalize since the 
    # the values in the matrix sum to 1
    kernel_sharpening = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    # applying different kernels to the input image
    img_processed = cv2.filter2D(img_scaled, -1, kernel_sharpening)
    
    # save image
    status = cv2.imwrite(destination,img_processed)
```
